```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pymatgen
#from pymatgen.analysis.phase_diagram import PatchedPhaseDiagram
from mp_api.client import MPRester


```

```python
from matminer.datasets import get_available_datasets
from matminer.datasets import load_dataset
dft= load_dataset('matbench_glass')
dft
```

```python
from pymatgen.core.composition import Composition
for i in range (len(dft['composition'])):
    #print (dft['composition'][i])
    comp= Composition(dft['composition'][i])
    a=dict(comp.get_el_amt_dict())
    #print(a)
    dft['composition'][i]= a
dft
```

```python
df1 = pd.DataFrame({'Al' : []})
df1
```

```python
for i in range (len(dft['composition'])):
    aa= dft['composition'][i]
    df2= pd.DataFrame(aa, index=[i])
    df1= df1.append(df2)


df1
```

```python
df1= df1.fillna(0)
df1
```

```python
newdf = df1.join(dft['gfa'], lsuffix='_caller', rsuffix='_other')
newdf
```

```python
newdf["gfa"] = newdf["gfa"].astype(int)
newdf
```

```python
print (newdf.columns.values)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
 
df_scaled = scaler.fit_transform(newdf.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=[
  'Al','Ni','B','Co','Cr','Si','Fe','V','Ge','Cu','Mo','P','Ru','Rh','Ba',
 'La','Be','Ca','Mg','Ga','Zn','Ce','Ag','Mn','Nb','Pd','C','W','Dy','Zr',
'Er','Au','N','F','Sn','Gd','Hf','O','Ta','Ho','Pt','Ti','In','Ir','Nd',
    'Pr','Sb','Pb','Sc','Sm','Sr','Tb','U','Y','Yb','gfa'])
```

```python
df_scaled
```

```python
y=df_scaled["gfa"]
x=df_scaled.drop(['gfa'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import KFold, cross_val_score


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

rfc= RandomForestClassifier(n_estimators = 500, random_state = 0)

rfc.fit(X_train, y_train)

k_folds = KFold(n_splits = 50)

scores = cross_val_score(rfc, x, y, cv = k_folds)

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
print('done')
```
